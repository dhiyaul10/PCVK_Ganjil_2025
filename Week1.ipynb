```python
## Langkah (Step 1):
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

```python
## Langkah (Step 2):
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placecats.com/800/571"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to RGB
  final_frame = cv.hconcat((image, image_2))      #concatenate image
  cv2_imshow(final_frame)                         #show image
  print('\n')
```

```python
## Langkah (Step 3):
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
## Langkah (Step 4):
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis hrizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]
  final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Konversi ke RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()

# Ambil ukuran gambar
tinggi, lebar = image_2.shape[:2]

# -----------------------------
# Membuat garis VERTIKAL di tengah image
for y in range(tinggi):
    image_3[y, lebar // 2] = [255, 255, 255]

# Gabungkan (RGB vs Vertikal)
final_frame1 = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame1)


# -----------------------------
# Salinan baru untuk garis silang
image_3 = image_2.copy()

# Membuat garis SILANG (X)
for x in range(lebar):
    y1 = round((tinggi - 1) * x / (lebar - 1))              # diagonal kiri-atas → kanan-bawah
    y2 = (tinggi - 1) - round((tinggi - 1) * x / (lebar-1)) # diagonal kiri-bawah → kanan-atas
    image_3[y1, x] = [255, 255, 255]
    image_3[y2, x] = [255, 255, 255]

# Gabungkan (RGB vs Silang)
final_frame2 = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame2)
```

```python
# Konversi ke RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()

# Ambil ukuran gambar
tinggi, lebar = image_2.shape[:2]

# -----------------------------
# Membuat garis HORIZONTAL di tengah dengan panjang tertentu
panjang = lebar // 2            # misalnya setengah lebar gambar
x_start = (lebar - panjang) // 2
x_end = x_start + panjang

y = tinggi // 2  # posisi garis di tengah secara vertikal

for x in range(x_start, x_end):
    image_3[y, x] = [255, 255, 255]

# Gabungkan hasil (kiri: RGB, kanan: garis horizontal)
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Konversi ke RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()

# Ambil ukuran gambar
tinggi, lebar = image_2.shape[:2]

# -----------------------------
# Tentukan posisi dan ukuran kotak
kotak_w, kotak_h = lebar // 4, tinggi // 4  # ukuran kotak
x1, y1 = lebar // 3, tinggi // 3            # posisi kiri-atas kotak
x2, y2 = x1 + kotak_w, y1 + kotak_h

# Isi kotak dengan warna putih
image_3[y1:y2, x1:x2] = [255, 255, 255]

# -----------------------------
# Gabungkan hasil (RGB vs dengan kotak putih)
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
